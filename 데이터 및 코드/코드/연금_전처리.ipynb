```python
df.columns=['자치구','특례노령연금수급자수','특례노령연금금액','노령연금(20년이상)수급자수','노령연금(20년이상)금액','노령연금(10년이상~20년미만)수급자수','노령연금(10년이상~20년미만)금액','조기노령연금수급자수','조기노령연금금액','분할노령연금수급자수','분할노령연금금액']
```

```python
df
```

```python
df.drop(df.index[0:4],axis=0,inplace=True)
```

```python
df
```

```python
df['특례노령연금수급자수'] = pd.to_numeric(df['특례노령연금수급자수'])
df['특례노령연금금액'] = pd.to_numeric(df['특례노령연금금액'])
df['노령연금(20년이상)수급자수'] = pd.to_numeric(df['노령연금(20년이상)수급자수'])
df['노령연금(20년이상)금액'] = pd.to_numeric(df['노령연금(20년이상)금액'])
df['노령연금(10년이상~20년미만)수급자수'] = pd.to_numeric(df['노령연금(10년이상~20년미만)수급자수'])
df['노령연금(10년이상~20년미만)금액'] = pd.to_numeric(df['노령연금(10년이상~20년미만)금액'])
df['조기노령연금수급자수'] = pd.to_numeric(df['조기노령연금수급자수'])
df['조기노령연금금액'] = pd.to_numeric(df['조기노령연금금액'])
df['분할노령연금금액'] = pd.to_numeric(df['분할노령연금금액'])
df['분할노령연금수급자수'] = pd.to_numeric(df['분할노령연금수급자수'])
print(df.dtypes)
```

```python
df["노령연금총금액"]=df["특례노령연금금액"]+df["노령연금(20년이상)금액"]+df["노령연금(10년이상~20년미만)금액"]+df["조기노령연금금액"]+df["분할노령연금금액"]
```

```python
df["노령연금총수급자수"]=df["특례노령연금수급자수"]+df["노령연금(20년이상)수급자수"]+df["노령연금(10년이상~20년미만)수급자수"]+df["조기노령연금수급자수"]+df["분할노령연금수급자수"]
```
