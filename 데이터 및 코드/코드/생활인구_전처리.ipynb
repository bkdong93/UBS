```python
import glob
import os

input_file= r'' #데이터 파일들이 있는 디렉토리 위치
output_file= r'2018_sorted.csv' #각 데이터 파일을 병합한 새로운 파일에 사용하고 싶은 이름

allFile_list= glob.glob(os.path.join(input_file, 'sorted_2018*')) #파일 이름이 'sorted_2018'로 시작하는 모든 파일 병합

print(allFile_list)

allData= [] #읽어 들인 파일 내용을 저장할 빈 리스트
for file in allFile_list:
    df= pd.read_csv(file) #데이터 파일들을 읽어내기
    allData.append(df) #빈 리스트에 읽어낸 내용 추가
    
dataCombine= pd.concat(allData, axis=0, ignore_index=True) #concat함수를 이용해서 리스트의 내용을 병합. 이 때 기존 순서를 무시하고 순서대로, 수직으로 병합

dataCombine.to_csv(output_file, index=False)
```

```python
#한 행정동 코드 내의 모든 생활인구(일자별, 시간별) 합치기
df1= data.groupby(['행정동코드'], as_index=False).mean()
```

```python
df1
```

```python
#일자와 시간은 행정동별로 합쳤으므로, 일자와 시간 컬럼 삭제

df1= df1.iloc[:, [0,3,4,5,6,7,8]]
df1
```

```python
#모든 성별의 60세 이상 생활인구를 구하기 위해 합해야 하기 때문에 area_code 컬럼을 제외하고 새 데이터프레임 생성
df2=df1.iloc[:, [1,2,3,4,5,6]]

df2
```

```python
#합을 구해 over_60이라는 컬럼으로 생성
df2[ 'over_60' ] = df2.sum(axis=1)

df2
```

```python
#df1과 df2에서 필요한 컬럼은 각각 area_code와 over_60
#df1과 df2를 merge하기 위해 공통된 컬럼이 필요해 m_60~64 컬럼만 남겨두고, 이외의 컬럼은 삭제

df1= df1.iloc[:, [0,1]]
df1
```

```python
df2= df2.iloc[:, [0,6]]
df2
```

```python
#m_60~64로 merge 후 필요없는 컬럼인 m_60~64를 제외 후 새 데이터프레임 생성
data= pd.merge(df1, df2, on="m_60~64", how='left')
data= data.iloc[:, [0,2]]

data
```

```python
#area_code의 데이터타입을 int로 변경

data= data.astype({'행정동코드': 'int'})

data
```

```python
data.info()
```

```python
#생활인구 원본데이터에서 일부 행정동 코드가 잘못 입력된 사실을 확인 후 수정

data= data.replace(to_replace= 11305590, value= 11305595, regex= True)
data= data.replace(to_replace= 11305600, value= 11305603, regex= True)
data= data.replace(to_replace= 11305606, value= 11305608, regex= True)
data= data.replace(to_replace= 11305610, value= 11305615, regex= True)
data= data.replace(to_replace= 11305620, value= 11305625, regex= True)
data= data.replace(to_replace= 11305630, value= 11305635, regex= True)
```
