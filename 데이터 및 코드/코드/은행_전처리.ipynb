```python
dong = pd.read_excel("geojson_행정동코드.xlsx") ## 누락된 행정동 이름 비교 후 입력
bfinal = pd.read_excel("test.xlsx")
```

```python
bfinal.rename(columns={'자치구':'구'}, inplace=True)
bfinal
```

```python
dong_area = dong["동"]
```

```python
banks = bfinal["동"].values
```

```python
dong_arr = [] ## 누락된 동 추가
for i in dong_area:
    if i not in banks:
        dong_arr.append(i)
```

```python
serarr = pd.Series(dong_arr) ## 배열을 series로 변경
```

```python
dis_arr = [] ## 누락된 동에 해당하는 구추가
for i in dong_area:
    if i not in banks:
        dis_arr.append(dong[dong["동"] == i]["구"].values[0])
```

```python
ser_dis_arr = pd.Series(dis_arr) ##배열을 series 변경
```

```python
for i in range(len(serarr)): ## 데이터 완성
    series = pd.DataFrame([[0]*bfinal.shape[1]], columns= bfinal.columns)
    series["동"] = serarr[i]
    series["구"] = ser_dis_arr[i]
    bfinal = bfinal.append(series, ignore_index=True)
```

```python
## 많이 사용하는 시중은행만 사용
bfinal.drop(labels=["Unnamed: 0","기간","합계","한국씨티은행","한국은행","SC제일은행","지방은행","KDB산업은행","수협은행","한국수출입은행","외국은행"], axis=1, inplace=True)
bfinal
```

```python
## 합계 만들기
bfinal["행정동별은행합계"] = bfinal["우리은행"] + bfinal["KB국민은행"] + bfinal["신한은행"] + bfinal["KEB하나은행"] + bfinal["IBK기업은행"] + bfinal["NH농협은행"]    
```
