```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.preprocessing import MinMaxScaler
```

```python
data= pd.read_excel("생활인구.xlsx", encoding='utf-8')
```

```python
#0~1 스케일링

scaler= MinMaxScaler()
data[["행정동별생활인구합계"]]= scaler.fit_transform(data[["행정동별생활인구합계"]])
data
```

```python
#xlsx 파일로 저장

data.to_excel("생활인구_정규화.xlsx", encoding='utf-8', index=False)
```

```python
low = data.sort_values(by = ['행정동별생활인구합계', '행정동코드'], ascending = True).head(10)
#low
```

```python
high = data.sort_values(by = ['행정동별생활인구합계', '행정동코드'], ascending = False).head(10)
#high
```
